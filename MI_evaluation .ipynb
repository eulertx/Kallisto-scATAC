Adjust Mutual Information (MI) between CellRanger and DHS500 count matrices 
for different number of regions considered for DHS500 

```python
###################
# Import packages #
###################
import scanpy as sc
import sklearn.metrics
from math import sqrt
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statistics
import warnings
warnings.filterwarnings('ignore')
```

```python
####################################
# CellRanger count matrix analysis #
####################################

adata_10x = sc.read_10x_h5("atac_v1_pbmc_10k_filtered_peak_bc_matrix.h5", gex_only=False)
print("\n10x", adata_10x)
```

```python
sc.pp.filter_cells(adata_10x, min_genes=200)
sc.pp.filter_genes(adata_10x, min_cells=10)

# Normalize the counts in the matrix
sc.pp.normalize_per_cell(adata_10x, counts_per_cell_after=1e4)
sc.pp.log1p(adata_10x)
adata_10x.raw = adata_10x

# Highly variable genes 
sc.pp.highly_variable_genes(adata_10x, min_disp=0.5, min_mean=0.1, inplace=True)
adata_10x  = adata_10x[:, adata_10x.var['highly_variable']]
```

```python
# Cluster
sc.pp.scale(adata_10x)
sc.tl.pca(adata_10x)

sc.pp.neighbors(adata_10x, n_pcs=10)
sc.tl.umap(adata_10x)

sc.tl.leiden(adata_10x, resolution = 0.2)
sc.pl.umap(adata_10x, color='leiden')

adata_10x.obs_names = [x.split('-')[0] for x in adata_10x.obs_names]
```

MI evaluation for different number of regions is possible both for DHS and DHS500.
Hereafter is presented the analysis for DHS500.

The MI is evaluated every time and the result recorded into a file.
The time required for this analysis is not small. It can be changed by modifying the sampling (n) 

```python
mutual_info=[]
n = np.arange(5000, 200000, 10000)
```

```python
for i in range(len(n)):
    print("\n")
    print("#########","n",n[i],"#########")
    adata_dhs = sc.read("counts_hg19_dhs_merged500bp.h5ad", gex_only=False)

    adata_select = adata_dhs[:,adata_dhs.var_names[:n[i]]]
    print(adata_select)
    
    # Normalize the counts in the matrix
    sc.pp.normalize_per_cell(adata_select, counts_per_cell_after=1e4)
    sc.pp.log1p(adata_select)
    adata_select.raw = adata_select
    
    # Highly variable genes
    sc.pp.highly_variable_genes(adata_select, min_disp=0.5, min_mean=0.1, inplace=True)
    adata_select = adata_select[:, adata_select.var['highly_variable']]
    
    # Cluster
    sc.pp.scale(adata_select)
    sc.tl.pca(adata_select)

    sc.pp.neighbors(adata_select, n_pcs=10)
    sc.tl.umap(adata_select)

    sc.tl.leiden(adata_select, resolution = 0.43)
    sc.pl.umap(adata_select, color='leiden')

    #########################
    #   MUTUAL INFORMATION  #
    #########################
    cl_10x_or = adata_10x.obs['leiden']
    cl_dhs_or = adata_select.obs['leiden']

    cl_10x_or.index = [x.split('-')[0] for x in cl_10x_or.index]

    common_cells_10x_dhs = set(cl_10x_or.index).intersection(set(cl_dhs_or.index))

    cl_10x = cl_10x_or.loc[common_cells_10x_dhs]
    cl_dhs = cl_dhs_or.loc[common_cells_10x_dhs]
    print("Mutual Information 10x-DHS500: ", sklearn.metrics.adjusted_mutual_info_score(cl_10x, cl_dhs))

    info = str(n[i]) + "\t" + str(sklearn.metrics.adjusted_mutual_info_score(cl_10x, cl_dhs)) + "\t" + str(adata_10x.shape[0]) + "\t" + str(adata_10x.shape[1]) + "\t" + str(adata_select.shape[0]) + "\t" + str(adata_select.shape[1])
    mutual_info.append(info)
```

```python
# file to record for each sampling n the MI anad the shape of both count matrices 
with open("MI_dhs500_10x.txt", "w") as f:
   for line in mutual_info:
        f.write("%s\n" % line)
```

This is the way used to obtained the MI plotted in Figure 1C of the manuscript.
